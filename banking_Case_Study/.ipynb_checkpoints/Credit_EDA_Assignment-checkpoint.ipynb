### Standardizing Values

#### DAYS_BIRTH, DAYS_EMPLOYED, DAYS_REGISTRATION, DAYS_ID_PUBLISH is in negative values. Convert it into positive values.

```python
df.DAYS_BIRTH.value_counts(normalize = True)
```

```python
#Converting DAYS_BIRTH to years. Column renamed to YEARS_BIRTH
df.DAYS_BIRTH=df.DAYS_BIRTH.apply(lambda x: abs(x//365.25))
df.rename({'DAYS_BIRTH':'YEARS_BIRTH'}, axis=1, inplace=True)
df.info()
```

```python
#DAYS_REGISTRATION, DAYS_ID_PUBLISH are converting to years. Renaming the columns


df.DAYS_REGISTRATION = df.DAYS_REGISTRATION.apply(lambda x: abs(x//365.25))
df.DAYS_ID_PUBLISH = df.DAYS_ID_PUBLISH.apply(lambda x: abs(x//365.25))
df.DAYS_EMPLOYED= df.DAYS_EMPLOYED.apply(lambda x: abs(x//365.25))
df.rename({'DAYS_REGISTRATION':'YEARS_REGISTRATION', 'DAYS_ID_PUBLISH':'YEARS_ID_PUBLISH', 'DAYS_EMPLOYED':'YEARS_EMPLOYED'}, axis=1, inplace=True)
```

```python
df.YEARS_REGISTRATION.value_counts(normalize = True)
```

### Creating buckets

```python
# Bucketting Years Birth variable and Creating new column "BIRTH_YEAR_GROUP"

df['BIRTH_YEAR_GROUP']= pd.cut(df.YEARS_BIRTH,bins=[20,30,40,50,60,70])

df['BIRTH_YEAR_GROUP'].value_counts(normalize=True)
```

```python
df['BIRTH_YEAR_GROUP'].value_counts(normalize=True).plot.barh()
plt.show()
```

```python
# Bucketting 'AMT_INCOME_TOTAL' and Creating new column "TOTAL_INCOME_SIZE"

df['TOTAL_INCOME_SIZE']= pd.qcut(df['AMT_INCOME_TOTAL'], q=[0,0.1,0.3,0.6,0.8,1],labels=['Lowest','Low','Medium','High','Highest'])

df['TOTAL_INCOME_SIZE'].value_counts(normalize=True)
```

```python
df['TOTAL_INCOME_SIZE'].value_counts(normalize=True).plot.barh()
plt.show()
```
